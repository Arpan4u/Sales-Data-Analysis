##Need to check which city has the Higher No of Sales

```python
all_data.head()
```

```python
#need to add a city column with .apply method
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split (' ')[1]
    
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+' '+ get_state(x))  #running a fuction to split purchase address
all_data.head()
```

```python
view=all_data.groupby('City').sum()
view
```

```python
import matplotlib.pyplot as plt

Cities = [city for city ,df in all_data.groupby('City')]
plt.bar(Cities,view['Sales'])
plt.xticks(Cities,rotation='vertical', size=8)
plt.ylabel('Sales in USD')
plt.xlabel('City')
plt.show()
```

What time should we display advertisement to maximize customer's buying product?


```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

```python
all_data['Hour'] =all_data['Order Date'].dt.hour
all_data['Minute'] =all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour ,df in all_data.groupby('Hour')]
plt.plot(hours,all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()

#according to this graph,we can put advertisement by 11am and 7pm.Sale is high then
```
